```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
kf = StratifiedKFold(n_splits=5)
```

```python
import numpy as np
```

```python
x = np.random.randint(1,1000,600)
y = np.random.randint(1,1000,600)
```

```python
x
```

```python
data = {
    'me': x,
    'my':y
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['kfold'] = -1
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
y = df['me']
```

```python
kf.split(X=df, y=y)
```

```python
y
```

```python
for f, (t_, v_) in enumerate(kf.split(X=df,y=y)):
    print('fold:',f)
    # print('training:', t_)
    print('validation:', v_)
    df.loc[v_,'kfold'] = f
```
